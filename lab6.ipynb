Wczytuję wyniki obliczeń programu pliku `duszek`, zapisane do pliku `raport_duszka.txt`

```python
E_values = []
L_values = []
avg_energy_system = []
avg_energy_duszek = []
avg_magnetization = []
temperatures = []

with open("raport_duszka.txt", "r") as file:
    for line in file:
        values = line.split(",\t")
        E_values.append(int(values[0]))
        L_values.append(int(values[1]))
        avg_energy_system.append(float(values[2]))
        avg_energy_duszek.append(float(values[3]))
        avg_magnetization.append(float(values[4]))
        temperatures.append(float(values[5]))

data_dict = {}
for L in set(L_values):
    data_dict[L] = {
        "E_values": [],
        "avg_energy_system": [],
        "avg_energy_duszek": [],
        "avg_magnetization": [],
        "temperatures": []
    }

for i, L in enumerate(L_values):
    data_dict[L]["E_values"].append(E_values[i])
    data_dict[L]["avg_energy_system"].append(avg_energy_system[i])
    data_dict[L]["avg_energy_duszek"].append(avg_energy_duszek[i])
    data_dict[L]["avg_magnetization"].append(avg_magnetization[i])
    data_dict[L]["temperatures"].append(temperatures[i])
```

```python
import matplotlib.pyplot as plt
for L in [10, 20, 40]:
    plt.title(f"Zależność średniej magnetyzacji od temperatury dla L={L}")
    plt.plot(data_dict[L]["temperatures"], data_dict[L]["avg_magnetization"])
    plt.xlabel("Temperatura")
    plt.ylabel("Średnia magnetyzacja")
    plt.show()
    plt.title(f"Zależność średniej energii układu od temperatury dla L={L}")
    plt.plot(data_dict[L]["temperatures"], data_dict[L]["avg_energy_system"])
    plt.xlabel("Temperatura")
    plt.ylabel("Średnia energia układu")
    plt.show()
```
